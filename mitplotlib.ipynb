```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline
```

**Matplotlib**

```python
x = np.arange(-10, 11)
y = x**2
plt.figure(figsize = (8, 3))
plt.title ("Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, "k--") #K Color
plt.plot(x, -y, "ro")
```

```python
plt.title("My First Plot")
plt.plot(x, y)
plt.plot(x, -1 * y)
plt.figure(figsize=(4, 2))
plt.show()
```

```python
plt.figure(figsize= (6,3))
plt.plot(x, y)
plt.title("Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(1, 2, 1)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['x^2', 'vertical line'])


plt.xlabel("X")
plt.ylabel("X^2")
plt.subplot(1, 2, 2)
plt.plot(x, -1 * y)
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-x^2', 'Horizontal line'])
```

```python
plt.figure(figsize=(12,4))
plt.title("My 2nd Plot")
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel("X")
plt.ylabel("X square")

plt.subplot(1, 2, 2)
plt.plot(x, -1 * y)
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2','Horizontal Line'])
plt.xlabel("X")
plt.ylabel("X^2")
```

```python
x = np.arange(-10, 11)
y = x**2
plt.figure(figsize=(8,6))

plt.subplot(2, 2, 1)
plt.plot([0, 0, 0], [-10, 0, 100], label = "line")
plt.plot(x, y, color = "y", label = "curve")
plt.legend()
plt.title("Plot1")
plt.xlabel("X")
plt.ylabel("Y")


plt.subplot(2, 2, 2)
plt.plot([0, 0, 0], [-10, 0, 100], label = "line")
plt.plot(x, y , color = "r", label = "curve")
plt.legend()
plt.title("Plot2")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2, 2, 3)
plt.plot([0, 0, 0], [-10, 0, 100], label = "line")
plt.plot(x, y , color = "g", label = "curve")
plt.legend()
plt.title("Plot3")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2, 2, 4)
plt.plot([0, 0, 0], [-10, 0, 100], label = "line")
plt.plot(x, y , color = "b", label = "curve")
plt.legend()
plt.title("Plot4")
plt.xlabel("X")
plt.ylabel("Y")
```
